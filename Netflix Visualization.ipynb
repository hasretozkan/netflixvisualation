# Netflix Visualization

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import MultiLabelBinarizer
from datetime import datetime as dt
from itertools import compress
```

## Reading Data

```python
df=pd.read_csv("./netflix_titles.csv")
df.head()
```

```python
df.info()
```

## Editing data

```python
df.drop("show_id",axis=1,inplace=True)
df.director.fillna("Unknown",inplace=True)
df.dropna(inplace=True)
df.rename({"listed_in":"genre"},inplace=True,axis=1)
df['country']=df.country.str.split(", ",expand=True)[0]
df[["duration","duration_type"]]=df.duration.str.split(" ",expand=True)
df=df.astype({"duration":"int64"})
df.genre=df.genre.apply(lambda x:x.split(", "))
mlb=MultiLabelBinarizer()
genre_mlb=pd.DataFrame(
    mlb.fit_transform(df.genre),
    columns=mlb.classes_,
    index=df.index).add_prefix("genre_")
df.drop("genre",axis=1,inplace=True)
df=pd.concat([df,genre_mlb],axis=1)
df.date_added=df.date_added.apply(lambda x:dt.strptime(x.strip(),"%B %d, %Y"))
df['year_added']=df.date_added.dt.year
df['month_added']=df.date_added.dt.month
df['day_added']=df.date_added.dt.month
df['release_added_diff']=abs(df.release_year-df.year_added)
```

```python
df.head()
```

## Visualization

```python
fig,axes=plt.subplots(figsize=(12,6))
plt.pie(df.type.value_counts(),explode=(0.05,0.05),colors=['#00aced','#ffa700'],autopct="%1.1f%%",labels=df.type.value_counts().index)
plt.title("Movies vs TV Shows")
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(16,8))
sns.histplot(ax=ax[0],x="duration",data=df.loc[df.type=="Movie",:],bins=50)
sns.histplot(ax=ax[1],x="duration",data=df.loc[df.type=="TV Show",:],bins=50)
```

```python
sns.countplot(y="country",hue="type",order=df.country.value_counts().index[0:10],data=df,palette="Set2")
```

```python
fig,ax=plt.subplots(figsize=(12,7))
sns.scatterplot(x=df.release_year,y=df.year_added,hue=df.type)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
sns.countplot(y="year_added",data=df,palette="Set2")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
sns.countplot(y="rating",data=df,palette="Set2")
```

```python
fig,ax=plt.subplots(figsize=(12,6))
sns.countplot(y="rating",data=df,palette="Set2",hue="type")
```
